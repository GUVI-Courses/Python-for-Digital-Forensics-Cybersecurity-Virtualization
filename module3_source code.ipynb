## Program 2: Getting File Metadata

### Objective
To retrieve metadata of a file.

### Theory
os.stat() provides file size and timestamps useful for investigations.

### Program

````python
# Getting metadata of a specific file
import os
import time

file_name = "sample.txt"

if os.path.exists(file_name):
    stats = os.stat(file_name)
    print("File Size:", stats.st_size, "bytes")
    print("Last Modified:", time.ctime(stats.st_mtime))
else:
    print("File does not exist")
````

### Output
```
File does not exist
```

## Program 3: Extracting File Timestamps

### Objective
To extract timestamps from a file.

### Theory
os.stat() retrieves timestamps and time.ctime() converts them to readable form.

### Program

````python
# Extracting timestamps from a file
import os
import time

file_name = "sample.txt"

if os.path.exists(file_name):
    stats = os.stat(file_name)

    print("Created:", time.ctime(stats.st_ctime))
    print("Modified:", time.ctime(stats.st_mtime))
    print("Accessed:", time.ctime(stats.st_atime))
else:
    print("File not found")
````

### Output
```
File not found
```

## Program 4: Creating a Simple Timeline

### Objective
To create a timeline of files based on modification time.

### Theory
Files are sorted chronologically to reconstruct activity.

### Program

````python
# Creating a basic file timeline
import os
import time

events = []

for file in os.listdir("."):
    if os.path.isfile(file):
        stats = os.stat(file)
        events.append((stats.st_mtime, file))

# Sort events by modification time
events.sort()

for event in events:
    print(time.ctime(event[0]), "-", event[1])
````

### Output
```
Mon Jan  1 09:00:00 2025 - log.txt
Mon Jan  1 09:30:00 2025 - report.txt
Mon Jan  1 10:15:00 2025 - evidence.jpg
```

## Program 5: Hashing Text Data

### Objective
To generate SHA-256 hash of text.

### Theory
Hashing creates a unique fingerprint of data.

### Program

````python
import hashlib

data = "Digital Forensics"

# Convert string to bytes
encoded_data = data.encode()

hash_value = hashlib.sha256(encoded_data).hexdigest()

print("SHA-256 Hash:", hash_value)
````

### Output
```
SHA-256 Hash: 5c1b...a9d3
```

## Program 6: Hash Sensitivity

### Objective
To show how small changes affect hash.

### Theory
Even minor input changes produce completely different hashes.

### Program

````python
import hashlib

data1 = "Digital Forensics"
data2 = "Digital Forensics!"

hash1 = hashlib.sha256(data1.encode()).hexdigest()
hash2 = hashlib.sha256(data2.encode()).hexdigest()

print("Hash 1:", hash1)
print("Hash 2:", hash2)
````

### Output
```
Hash 1: 5c1b...a9d3
Hash 2: 8e7f...c2b1
```

## Program 7: Hashing a File

### Objective
To generate hash of a file.

### Theory
Binary data is hashed for integrity verification.

### Program

````python
import hashlib

file_name = "sample.txt"

try:
    with open(file_name, "rb") as file:
        content = file.read()
        hash_value = hashlib.sha256(content).hexdigest()
        print("File SHA-256:", hash_value)
except FileNotFoundError:
    print("File not found")
````

### Output
```
File not found
```

## Program 9: Chunk-Based Hashing

### Objective
To hash large files efficiently.

### Theory
Reads file in chunks to avoid memory issues.

### Program

````python
import hashlib

file_name = "sample.txt"
hash_obj = hashlib.sha256()

try:
    with open(file_name, "rb") as file:
        while chunk := file.read(4096):
            hash_obj.update(chunk)

    print("SHA-256:", hash_obj.hexdigest())
except FileNotFoundError:
    print("File not found")
````

### Output
```
File not found
```

## Program 10: Reading Email Headers

### Objective
To extract email header information.

### Theory
Python parses .eml files to retrieve metadata.

### Program

````python
import email
from email import policy
from email.parser import BytesParser

try:
    with open("sample_email.eml", "rb") as file:
        msg = BytesParser(policy=policy.default).parse(file)

        print("From:", msg["From"])
        print("To:", msg["To"])
        print("Subject:", msg["Subject"])
        print("Date:", msg["Date"])
except FileNotFoundError:
    print("Email file not found")
````

### Output
```
Email file not found
```

## Program 11: Extracting Received Headers

### Objective
To trace email route.

### Theory
Received headers show path of email servers.

### Program

````python
import email
from email import policy
from email.parser import BytesParser

try:
    with open("sample_email.eml", "rb") as file:
        msg = BytesParser(policy=policy.default).parse(file)

        for header in msg.get_all("Received", []):
            print(header)
except FileNotFoundError:
    print("Email file not found")
````

### Output
```
Email file not found
```

## Program 12: Extracting Email Body

### Objective
To read email content.

### Theory
Extracts plain text from email.

### Program

````python
import email
from email import policy
from email.parser import BytesParser

try:
    with open("sample_email.eml", "rb") as file:
        msg = BytesParser(policy=policy.default).parse(file)

        if msg.is_multipart():
            for part in msg.walk():
                if part.get_content_type() == "text/plain":
                    print(part.get_content())
        else:
            print(msg.get_content())
except FileNotFoundError:
    print("Email file not found")
````

### Output
```
Email file not found
```

## Program 13: Extracting Email Attachments

### Objective
To extract attachments from email.

### Theory
Attachments are decoded and saved.

### Program

````python
import email
from email import policy
from email.parser import BytesParser

try:
    with open("sample_email.eml", "rb") as file:
        msg = BytesParser(policy=policy.default).parse(file)

        for part in msg.iter_attachments():
            filename = part.get_filename()
            if filename:
                with open(filename, "wb") as f:
                    f.write(part.get_payload(decode=True))
                print("Saved attachment:", filename)
except FileNotFoundError:
    print("Email file not found")
````

### Output
```
Email file not found
```
