## Program 2: Hash Generation

### Objective
Generate hash

### Theory
Integrity check

### Program

````python
import hashlib

file_name = "sample.txt"

with open(file_name, "rb") as file:
    content = file.read()
    hash_value = hashlib.sha256(content).hexdigest()

print("File Hash:", hash_value)
````

### Output
```
3a6eb0790f39ac87c94f3856b2dd2c5d110e6811602261a9a923d3bb23adc8b7
```

## Program 4: CSV Report

### Objective
Create CSV

### Theory
Structured output

### Program

````python
import csv

data = [
    ["File Name", "Size", "Hash"],
    ["file1.txt", "1024", "abc123"],
    ["file2.txt", "2048", "def456"]
]

with open("report.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("CSV report generated")
````

### Output
```
report.csv file created with headers: File, Status
```

## Program 12: Monitoring Script

### Objective
