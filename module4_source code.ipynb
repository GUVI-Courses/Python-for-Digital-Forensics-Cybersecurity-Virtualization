## Program 1: Reading a Memory Dump File

### Objective
To read binary data from a memory dump.

### Theory
Python reads raw memory dump data in binary format.

### Program

````python
file_name = "memory_dump.raw"

try:
    with open(file_name, "rb") as file:
        data = file.read(64)
        print(data)
except FileNotFoundError:
    print("Memory dump file not found")
````

### Output
```
Memory dump file not found
```

## Program 2: Searching Memory for Patterns

### Objective
To identify patterns like IP addresses in memory.

### Theory
Regex helps locate suspicious patterns in memory data.

### Program

````python
import re

data = b"Connection to 192.168.1.50 established"

ip_pattern = rb"\d{1,3}(?:\.\d{1,3}){3}"

match = re.search(ip_pattern, data)

if match:
    print("IP Address Found:", match.group())
````

### Output
```
IP Address Found: b'192.168.1.50'
```

### Objective

### Theory

### Program


## Program 4: Reading Registry Values

### Objective
To extract startup programs from registry.

### Theory
winreg accesses Windows registry.

### Program

````python
import winreg

key = winreg.OpenKey(winreg.HKEY_CURRENT_USER,
                     r"Software\Microsoft\Windows\CurrentVersion\Run")

i = 0
try:
    while True:
        value = winreg.EnumValue(key, i)
        print(value)
        i += 1
except OSError:
    pass
````

### Output
```
('OneDrive', 'C:\\Program Files\\Microsoft\\OneDrive\\OneDrive.exe', ...)
```

## Program 5: Capturing Network Packets

### Objective
To capture live packets using Scapy.

### Theory
sniff() captures network packets.

### Program

````python
from scapy.all import sniff

def packet_callback(packet):
    print(packet.summary())

sniff(count=5, prn=packet_callback)
````

### Output
```
Ether / IP / TCP 192.168.1.5 → 142.250.183.78
```

## Program 8: Extracting IP Addresses

### Objective
To extract source and destination IPs.

### Theory
IP layer used to extract addresses.

### Program

````python
from scapy.all import sniff, IP

def packet_callback(packet):
    if IP in packet:
        print(packet[IP].src, "→", packet[IP].dst)

sniff(count=5, prn=packet_callback)
````

### Output
```
192.168.1.5 → 8.8.8.8
```

## Program 9: Reading Packets from PCAP

### Objective
To read packets from a capture file.

### Theory
rdpcap loads packet capture data.

### Program

````python
from scapy.all import rdpcap

packets = rdpcap("capture.pcap")

for packet in packets[:5]:
    print(packet.summary())
````

### Output
```
TCP 192.168.1.5 → 93.184.216.34
```

## Program 10: Extracting Network Sessions

### Objective
To identify communication sessions.

### Theory
Sessions are unique source-destination pairs.

### Program

````python
from scapy.all import rdpcap, IP

packets = rdpcap("capture.pcap")

sessions = set()

for packet in packets:
    if IP in packet:
        session = (packet[IP].src, packet[IP].dst)
        sessions.add(session)

for session in sessions:
    print("Session:", session)
````

### Output
```
Session: ('192.168.1.5', '8.8.8.8')
```

## Program 11: Detecting Suspicious Ports

### Objective
To detect unusual network ports.

### Theory
Uncommon ports may indicate suspicious activity.

### Program

````python
from scapy.all import rdpcap, TCP

packets = rdpcap("capture.pcap")

for packet in packets:
    if TCP in packet:
        port = packet[TCP].dport
        if port not in [80, 443, 53]:
            print("Suspicious Port:", port)
````

### Output
```
Suspicious Port: 4444
```
