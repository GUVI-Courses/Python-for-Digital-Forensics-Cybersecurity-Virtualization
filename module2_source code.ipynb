## Program 18: Else Finally

### Objective
Complete handling

### Theory
Flow control

### Program

````python
try:
    x = 10
    y = 2
    print(x / y)
except:
    print("Error occurred")
else:
    print("Operation successful")
finally:
    print("Execution completed")
````

```
Execution completed
```

## Program 23: Extract Digits

### Objective
Find numbers

### Theory
Regex digits

### Program

````python
import re

text = "User ID: 4521"
result = re.search(r"\d+", text)

print(result.group())
````

### Output
```
4521
```

## Program 24: Regex Groups

### Objective
Extract data

### Theory
Capture group

### Program

````python
import re

log = "Error code: 404"
match = re.search(r"(\d+)", log)

print(match.group(1))
````

### Output
```
404
```

## Program 25: Email Validation

### Objective
Validate email

### Theory
Regex pattern

### Program

````python
import re

email = "user@example.com"
pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"

print(re.match(pattern, email) is not None)
````

### Output
```
True
```

## Program 26: Regex File Search

### Objective
Search file

### Theory
Filter logs

### Program

````python
import re

with open("log.txt", "r") as file:
    for line in file:
        if re.search("error", line):
            print(line.strip())
````

### Output
```
Error: File not found
```
