## Program 2: File Integrity Verification

### Objective
Generate hash

### Theory
Verify file integrity

### Program

````python
import hashlib

file_name = "suspicious_file.exe"

with open(file_name, "rb") as file:
    hash_value = hashlib.sha256(file.read()).hexdigest()

print("SHA256:", hash_value)
````

### Output
```
SHA256: a3f5c7d9e1b2...
```

## Program 7: Generate File Hash

### Objective
Hash file

### Theory
Integrity check

### Program

````python
import os
import hashlib

base_path = os.path.dirname(__file__)
file_name = os.path.join(base_path, "evidence_data", "sample.txt")

with open(file_name, "rb") as file:
    file_hash = hashlib.sha256(file.read()).hexdigest()

print("File Hash:", file_hash)
````


## Program 8: Detect Suspicious Files

### Objective
Find malicious files

### Theory
Extension filtering

### Program

````python
import os

base_path = os.path.dirname(__file__)
evidence_folder = os.path.join(base_path, "evidence_data")

files = os.listdir(evidence_folder)

suspicious_extensions = [".exe", ".bat"]

for file in files:
    if file.endswith(tuple(suspicious_extensions)):
        print("⚠️ Suspicious file:", file)
````

### Output
```
⚠️ Suspicious file: malware.exe
```

## Program 11: Bar Chart

### Objective

### Program

````python
import matplotlib.pyplot as plt

file_types = ["PDF", "DOCX", "EXE", "ZIP"]
counts = [15, 10, 3, 7]

plt.bar(file_types, counts)

plt.title("Suspicious File Types Detected")
plt.xlabel("File Type")
plt.ylabel("Count")

plt.show()
````
