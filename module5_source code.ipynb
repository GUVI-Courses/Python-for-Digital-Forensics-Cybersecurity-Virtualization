## Program 1: Connecting to Virtual Machine Manager

### Objective
To connect to virtualization system using Python.

### Theory
libvirt allows Python to interact with virtualization platforms.

### Program

````python
import libvirt

conn = libvirt.open('qemu:///system')

if conn is None:
    print("Failed to connect to hypervisor")
else:
    print("Connected to virtualization system")
````

### Output
```
Connected to virtualization system
```

## Program 2: Listing Virtual Machines

### Objective
To list available virtual machines.

### Theory
Python retrieves VM names using libvirt.

### Program

````python
import libvirt

conn = libvirt.open('qemu:///system')

domains = conn.listAllDomains()

for domain in domains:
    print("VM Name:", domain.name())
````

### Output
```
VM Name: TestVM
VM Name: UbuntuVM
```

## Program 3: Starting Virtual Machine

### Objective
To start a virtual machine.

### Theory
Python can control VM execution.

### Program

````python
domain = conn.lookupByName("TestVM")

domain.create()
print("Virtual machine started")
````

### Output
```
Virtual machine started
```

## Program 4: Accessing Cloud Data Using Python

### Objective
To retrieve cloud storage data.

### Theory
boto3 allows Python to interact with AWS services.

### Program

````python
import boto3

s3 = boto3.client('s3')

response = s3.list_buckets()

for bucket in response['Buckets']:
    print(bucket['Name'])
````

### Output
```
forensics-data-bucket
logs-storage
backup-bucket
```
